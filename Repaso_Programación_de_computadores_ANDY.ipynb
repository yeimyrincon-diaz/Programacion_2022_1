###3. Haga un algoritmo que codifique el mensaje que ingresa el usuario por un aumento en cada caracter de su valor ascii,
El usuario ingresa tanto el mensaje como la clave de codificaci√≥n.


```python
#Recuerde la funcion ord() y chr(), para ir de caracter a ascii y viceversa
mensaje = input('Ingrese el mensaje: ') #esto ya es una cadena (arreglo) de caracteres
clave = int(input('Ingrese la clave de codificaci√≥n: '))
solucion=[]
for indice in range(0,len(mensaje),1): #range(start 0, stop en el tama√±o del mensaje, step de a 1). EL rango crea una lista de cero a n numeros
    #print(mensaje[i]) Llover de yo viera jajajajaja
    #indice son los indices
    # for hace que INDICE tenga el valor de cada uno de los elementos del rango
    ##print(indice)
    cambioaascii = ord(mensaje[indice]) + clave
    ##print(mensaje[indice],ord(mensaje[indice]))
    ##print(cambioaascii)
    cambioaascii = chr(cambioaascii)
    solucion.append(cambioaascii) # append a√±ade el caracter al arreglo de soluci√≥n
    print(solucion) #imprime la respuesta secuenccial de cada iteraci√≥n, porque estamos en una iteraci√≥n
    #indice es igual a la direci√≥n de cada casa, cada casa es una letra
    #print(i) # h esta en la direcci√≥n 0, h esta guardada en mensaje, las direcciones de mensaje estan guardadas en i
#print(chr(97))
#print(ord('a'))#al valor ascii hay que sumarle el valor de la clave ingresado por el usuario
```
